
### Innleveringsfrist:

* 13.03.2026 - 23:59

---
## Oppgave 1 - Medisinsk testing

En ny test for en sjelden sykdom er utviklet. Vi får følgende opplysninger:

__Begivenheter:__
* La $S$ betegne begivenheten med alle personer i befolkningen som har sykdommen.
* La $T$ betegne begivenheten med personer i befolkningen som vil teste positivt.

__Informasjon:__
* Sykdommen rammer $1\%$ av befolkningen.
* Testen gir positivt utslag for $95\%$ av alle personer som har sykdommen (sann positiv rate).
* Testen gir positivt utslag for $2\%$ av alle personer som ikke har sykdommen (falsk positiv rate).

**Spørsmål:** Hvis en person tar testen og får et positivt resultat, hva er sannsynligheten for at vedkommende faktisk har sykdommen?

**Bruk informasjonen over til først å angi med desimaltall:**
* $P(S) = ....$ (Andelen syke i befolkningen)
* $P(S^c) = ....$ (Andelen friske i befolkningen)
* $P(T|S) = ....$ (Sann positiv rate)
* $P(T|S^c) = ....$ (Falsk positiv rate)

--

**Du skal:** 
1. Lage et sannsynlighetstre som er tilpasset opplysningene i oppgaven. 
2. Beregne $P(S|T)$ - dvs sannsynligheten for at en person har sykdommen gitt at vedkommende tester positivt.

__Merk:__<br>
Løsningen av 2. krever bruk av Bayes' regel, så du bør først beregne $P(T)$. 

---


---
## Oppgave 2 - Terningspill med 3 rettferdige terninger
(Se tips til løsning nederst i oppgaven.)

### 2a) 
Vi kaster 3 rettferdige terninger og studerer følgende begivenheter:
- **Begivenhet $A$**: *Første terning viser et partall* (dvs. 2, 4 eller 6).  
- **Begivenhet $B$**: *Summen av de tre terningene er et oddetall*.  

**Oppgaver:**

1. Beskriv utfallsrommet $S$ for kast av 3 terninger (med antall utfall og hvordan de kan representeres).  
2. Bestem antall utfall i $A$ og antall utfall i $B$.  
3. Regn ut $P(A)$ og $P(B)$.  
4. Regn ut $P(A \cap B)$.  
5. Sjekk om $A$ og $B$ er uavhengige ved å undersøke om  
   $$ P(A \cap B) \stackrel{?}{=} P(A)\,P(B).$$

---

### 2b)
- **Begivenhet $C$**: *Summen av de tre terningene er lik 10*.  
- **Begivenhet $D$**: *Alle terningene viser forskjellige verdier*.  

**Oppgaver:**

1. Beskriv hvilke utfall som tilhører $C$ (alle kombinasjoner av tre terninger som summerer til 10).  
2. Beskriv hvilke utfall som tilhører $D$ (alle kombinasjoner der de tre terningene er forskjellige).  
3. Regn ut $P(C)$ og $P(D)$.  
4. Finn $P(C \cap D)$, altså sannsynligheten for at summen er 10 og alle terningene viser forskjellig verdi.  
5. Bestem den betingede sannsynligheten $P(D|C)$.
6. Undersøk uavhengighet ved å se om  
   $$P(C \cap D) \stackrel{?}{=} P(C) \, P(D).$$
7. Kan spørsmålet om uavhengighet også forklares fra den betingede sannsynligheten $P(D|C)$?

---

### Tips til løsning
- Et utfall av tre terningskast kan beskrives med et tall-trippel $(x_1,x_2,x_3)$, der koordinatverdiene $x_1, x_2, x_3\in \{1,2,3,4,5,6\}$.  
- For å finne sannsynligheter, bruk  
  $$
  P(\text{begivenhet}) = \frac{\text{antall utfall i begivenheten}}{\text{antall utfall i utfallsrommet}}.
  $$
- For å avgjøre om to begivenheter $A$ og $B$ er uavhengige, sjekk om  
  $$ P(A \cap B) = P(A)P(B).$$
  Hvis likheten ikke holder, er begivenhetene avhengige.

---

---

## Oppgave 3 - Julia-kode for Monte-Carlo simulering av sannsynligheter
### 3a)
Skriv et Julia-program som simulerer $n = 100.000$ kast av 3 terninger, og som estimerer sannsynligheten for de to begivenhetene 
* $A\cap B$ fra oppgave 2a), og
* $C\cap D$ fra oppgave 2b).

### 3b)
Utvid programmet i 3a) til også å estimere den betingede sannsynligheten $P(D\mid C)$.

__Husk:__<br>
Gi gode kommentarer som forklarer hvordan koden i programmet ditt virker, og sammenlign simuleringsresultatene med svarene du fant i Oppgave 2.

\
\
Lykke til!
